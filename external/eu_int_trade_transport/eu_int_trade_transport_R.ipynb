### Figure 1: EU Trade in Transport Services with extra EU, 2010-2017

* Here I pull the dataset I need (using the Eurostat Data Explorer). 'bop_its6_det' is the code to pull the International trade in services (since 2010) dataset. 

* I can choose filters to extract the exact data I need. The filter codes can be found on the Eurostat Data Explorer by clicking on the '+'.

* First I am going to assign the three filter codes I need to pull from the Eurostat dataset to stk_flow_var, so I can use it when I specify the filters I need in my dataset. 



```r
stk_flow_var <- c("CRE", "DEB", "BAL")
```

```r
dataset_fig1 <- get_eurostat_data(id="bop_its6_det", 
filters = list(bop_item="SC", currency="MIO_EUR", partner="EXT_EU28",geo="EU28",stk_flow=stk_flow_var),
date_filter = "2010:2017")
```

* dataset_fig1$time calls the 'time' variable in the dataset and changes the years from characters to numeric values

* NB: This is important for line graphs 

```r
dataset_fig1$time<-as.numeric(as.character(dataset_fig1$time))
```

* Calls the 'values' variable and divides each of the values by 1000 so that I can make the y-axis of the graph "EUR 1000 million"

```r
dataset_fig1$values<-dataset_fig1$values/1000
```

* Allows you to get a quick sneak-peak of the data you've pulled (to see the whole dataset, just take out the 'head' and the brackets. 

```r
head(dataset_fig1)
```

* Here I'm changing the order of the factor levels so I can make sure the legend on the graph is in the correct order

```r
dataset_fig1$stk_flow <- factor(dataset_fig1$stk_flow, levels = c("CRE", "DEB", "BAL"))
```

* The line below allows you to control the size of your graph

```r
options(repr.plot.width=10, repr.plot.height=5)
```

### Line graph - Figure 1

* ggplot2 allows you to visualise your data pretty efficiently. The first line calls the specific dataset you are using (the one you pulled using get_eurostat_data above!) and assigns the axes through 'aes'. The second line turns the graph into a line graph.

* scale_y_continuous and scale_x_continuous: control the scales of your y and x axes.

* ylab and xlab: control the labels for your axes. 

* ggtitle: gives your graph a title. 

* Your colours and legend labels can be added through scale_colour_manual - for my colours, I used a digital colour meter to extract the exact same sRGB colour codes used on the Statistics Explained website, but you can also simply replace these with "blue" or "green" and so on - as long as you're not too creative with your colour choices! 

```r
figure1 <- ggplot(data=dataset_fig1, aes(x=time, y=values, group=factor(stk_flow), color=factor(stk_flow))) + 
                      
geom_line() +

scale_color_manual(labels = c("Exports", "Imports", "Balance"), 
values = c("#D73C41", "#276EB4", "#B9C31E"), name = " ") +
ggtitle("Figure 1: EU trade in transport services with extra EU, 2010-2017") +

scale_y_continuous(limits = c(0, 180), breaks = seq(0, 180, by = 20)) +
scale_x_continuous(limits = c(2010, 2017), breaks = seq(2010, 2017, by = 1)) +
  
ylab("(EUR 1 000 million)") +
xlab(" ")  

print(figure1)
```

### Figure 2 - Contributions of sea, air, other modes of transport as well as postal and courier services to exports of EU-28 with extra-EU, 2010-2017 (EUR 1 000 million) 



* Pulling the relevant data for this chart in the same way as I did in Figure 1

* I start, once again, by assigning the four filter codes I need to 'bop', so that I can easily specify all four in the filters list. 

```r
id="bop_its6_det"
bop <- c("SC1", "SC2", "SC4", "SC3")

dataset_fig2 <-  get_eurostat_data(id, 
filters = list(bop_item=bop, currency="MIO_EUR", partner="EXT_EU28",geo="EU28",
stk_flow="CRE"), date_filter = "2010:2017")
```

* Calls the 'values' variable and divides each of the values by 1000 so that I can make the y-axis of the graph "EUR 1000 million"

```r
dataset_fig2$values<-dataset_fig2$values/1000
```

* A quick look at the dataset we just made

```r
head(dataset_fig2)
```

* Changing the order of the factor levels so I can make sure the values are in the correct order for the stacked bar chart

```r
dataset_fig2$bop_item <- factor(dataset_fig2$bop_item, levels = c("SC4", "SC3", "SC2", "SC1"))
```

### Stacked bar chart: Figure 2

```r
options(repr.plot.width=10, repr.plot.height=5)

figure2 <- ggplot(data=dataset_fig2, aes(x = time, y = values, fill = bop_item)) +
geom_bar(stat="identity")+scale_fill_manual(values = c("#71A8DF","#276EB4", 
"#E78A8D", "#D73C41"), labels = c("Postal and courier services", "Other modes of 
transport", "Air transport", "Sea transport"), name = " ") +

ggtitle("Figure 2: Contributions of sea, air, other modes of transport as 
well as postal and courier services to exports of EU-28 with extra-EU, 2010-2017") +

scale_y_continuous(limits = c(0, 180), breaks = seq(0, 180, by = 20))  +

ylab("(EUR 1 000 million)") +
xlab(" ")  

print(figure2)
```

### Figure 3 "Contributions of sea, air, other modes of transport as well as postal and courier services to imports of EU-28 with extra-EU, 2010-2017"

* The same as code Figure 2 but visualising imports instead of exports (stk_flow "CRE" changed to "DEB")

```r
id="bop_its6_det"

bop <- c("SC1", "SC2", "SC4", "SC3")

dataset_fig3 <-  get_eurostat_data(id, 
filters = list(bop_item=bop, currency="MIO_EUR", partner="EXT_EU28",geo="EU28",
stk_flow="DEB"), date_filter = "2010:2017")

dataset_fig3$values<-dataset_fig3$values/1000

dataset_fig3$bop_item <- factor(dataset_fig3$bop_item, 
levels = c("SC4", "SC3", "SC2", "SC1"))

options(repr.plot.width=10, repr.plot.height=5)

figure3 <- ggplot(data=dataset_fig3, aes(x = time, y = values, fill = bop_item)) +
geom_bar(stat="identity")+scale_fill_manual(values = c("#71A8DF","#276EB4", 
"#E78A8D", "#D73C41"), labels = c("Postal and courier services", "Other modes of transport", 
"Air transport", "Sea transport"), name = " ") +

ggtitle("Figure 3: Contributions of sea, air, other modes of transport as well as postal 
and courier services to imports of EU-28 with extra-EU, 2010-2017") +

scale_y_continuous(limits = c(0, 180), breaks = seq(0, 180, by = 20))  +

ylab("(EUR 1 000 million)") +
xlab(" ")  

print(figure3)
```
